```python
import sklearn as sk
import pandas as pd
import xgboost
from sklearn.model_selection import train_test_split
```

```python
train_file_paths = [f'36_TrainingData/L{i}_Train.csv' for i in range(1, 18)]#讀取全部訓練資料
df_list = [pd.read_csv(file) for file in train_file_paths]
df_data = pd.concat(df_list, ignore_index=True)
df_data
```

```python
i = input('請輸入要讀取的資料編號(1~17):')
while i.isdigit() == False or int(i) < 1 or int(i) > 17:
    i = input('輸入錯誤，請重新輸入:')
    
train_file_paths_single = f'36_TrainingData/L{i}_Train.csv'#讀取單一資料
df_analysised_data = pd.read_csv(train_file_paths_single)
print(f'已讀取第{i}筆資料')
```

```python
df_train, test, df_y_train, y_test = train_test_split(
    df_data[['WindSpeed(m/s)','Pressure(hpa)','Temperature(°C)','Humidity(%)','Sunlight(Lux)']], 
    df_data['Power(mW)'], test_size=(df_data.shape[0] - df_analysised_data.shape[0]) / df_data.shape[0])
df_a_train, df_a_y_train = df_analysised_data[['WindSpeed(m/s)','Pressure(hpa)','Temperature(°C)','Humidity(%)','Sunlight(Lux)']], df_analysised_data['Power(mW)']
```

```python
df_xgb = xgboost.XGBRegressor()
df_a_xgb = xgboost.XGBRegressor()
```

```python
df_xgb.fit(df_train, df_y_train)
df_a_xgb.fit(df_a_train, df_a_y_train)
```

```python
print(df_xgb.score(df_train, df_y_train))
print(df_a_xgb.score(df_a_train, df_a_y_train))
```

```python
print(df_xgb.score(test, y_test))
print(df_a_xgb.score(test, y_test))
```

```python
print(df_xgb.score(df_a_train, df_a_y_train))
print(df_a_xgb.score(df_train, df_y_train))
```
